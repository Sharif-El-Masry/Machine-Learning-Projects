```python
import matplotlib.pyplot as plt
import seaborn as sns

# Annahme: df ist dein DataFrame mit den Daten
fig = plt.figure(figsize=(16, 8))
cor = df.corr()
sns.heatmap(cor, annot=True, cmap=plt.cm.Reds)

# Anzeige der Figur auf dem Bildschirm
plt.show()

# Speichern der Figur als PNG
fig.savefig('heatmap.png', dpi=300)
```

```python
#visualization of numeric variables
numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns
for column in numeric_columns:
    plt.figure(figsize=(8, 6))
    # Histogramm fuÌr IsBadBuy = 0 (nicht schlechter Kauf)
    plt.hist(df.loc[df['IsBadBuy'] == 0, column], bins='auto', alpha=0.5, color='blue', label='Nicht schlechter Kauf')
    # Histogramm fuÌr IsBadBuy = 1 (schlechter Kauf)
    plt.hist(df.loc[df['IsBadBuy'] == 1, column], bins='auto', alpha=0.5, color='red', label='Schlechter Kauf')
    plt.xlabel(column)
    plt.ylabel('Anzahl')
    plt.title(f'Histogramm von {column}')
    plt.legend()
```

```python
#EFA: outlier
numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns
for column in numeric_columns:
    plt.figure(figsize=(8, 6))
    plt.hist(df[column], bins='auto')
    plt.xlabel(column)
    plt.ylabel('Anzahl')
    plt.title(f'Histogramm von {column}')
    plt.show()
```
