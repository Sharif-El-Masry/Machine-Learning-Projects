```python
def change_data_type(df): 
    """Returns cleaned DataFrame.
    Transform datatypes:
    -transform 'pickup_datetime' to datetime format
    -transforms every object type to category type
    Args:
    df (pd.DataFrame) : uncleaned DataFrame
    Returns:
    df (pd.DataFrame) : cleaned DataFrame
    """
    
    #to datetime
    df['PurchDate'] = pd.to_datetime(df['PurchDate'])  # Direct column access
    # Iteriere uÌber jede Spalte im DataFrame
    for column in df.columns:
        # ÃberpruÌfe, ob der Datentyp der Spalte Object ist
        if df[column].dtype == 'object':
            # Wandle den Datentyp in Category um
            df[column] = df[column].astype('category')
    # 'WheelTypeID' 'BYRNO' 'IsOnlineSale' sind alle 3 als int oder float klassifiziert, muÌssen aber category sein
    df['WheelTypeID'] = df['WheelTypeID'].astype('category')  # Direct column access
    df['BYRNO'] = df['BYRNO'].astype('category')  # Direct column access
    df['IsOnlineSale'] = df['IsOnlineSale'].astype('category')  # Direct column access
    
    return df 

df = change_data_type(df)
```

```python
def impute_metric_na(features_train):
       
    # Berechnen der Medianwerte fÃ¼r jede Spalte von features_train
    
    median_values = features_train.median(numeric_only=True)
    # Ersetzen der fehlenden Werte in features_train mit den berechneten Medianwerten
    features_train_imputed = features_train.fillna(value=median_values)
    
    return features_train_imputed, median_values


```

```python
import pandas as pd

def impute_missing_with_mode_and_fillna_aucguart_primeunit(df):
    """
    Ersetzt fehlende Werte in den Spalten mit dem Datentyp "category" mit den Modi.
    Args:
    df (pd.DataFrame): Der DataFrame, dessen Modi verwendet werden sollen.

    Returns:
    pd.DataFrame: Der DataFrame mit den ersetzen Werten.
    """
    
    # Eine Hilfsfunktion, um fehlende Werte in den angegebenen Spalten zu ersetzen
    def fill_missing_values(df, columns, values):
        for column in columns:
            df[column] = df[column].cat.add_categories([values[column]])
            df[column].fillna(values[column], inplace=True)
            
    # Ersetze fehlende Werte in den spezifizierten Spalten mit 'Unbekannt'
    fill_missing_values(df, ['PRIMEUNIT', 'AUCGUART', 'WheelType'], {'PRIMEUNIT': 'Unbekannt', 'AUCGUART': 'Unbekannt', 'WheelType': 'Unbekannt'})
    
    # Berechne die Modi des DataFrames
    modes = df.mode().iloc[0]
    
    # Ersetze fehlende Werte in allen kategorialen Spalten von df mit den Modi
    for column in df.select_dtypes(include='category').columns:
        df[column].fillna(value=modes[column], inplace=True)

    return df, modes
```

```python
#pca
from sklearn.decomposition import PCA
def apply_pca(train_df, selected_columns, n_components=1):
    # Extraktion der relevanten Spalten aus dem Trainingsdatensatz
    train_data = train_df[selected_columns].copy()
   
    # Initialisierung der PCA
    pca = PCA(n_components=n_components)
    # Anwendung der PCA auf die Trainingsdaten
    transformed_train_data = pca.fit_transform(train_data)
    
    # HinzufuÌgen der neuen Variable zum Trainingsdatensatz
    train_df['PCA_Component'] = transformed_train_data[:, 0]
    
    # Entfernen der urspruÌnglichen Spalten aus dem Trainingsdatensatz
    train_df.drop(columns=selected_columns, inplace=True)
    
    
    return train_df, pca




```
